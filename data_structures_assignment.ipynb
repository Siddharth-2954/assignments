```python
# Task - 4
```

```python
catalog = []
```

```python
for product in products:
    if product in price_dict:
        index = products.index(product)
        category = categories[index]
        catalog.append((product, price_dict[product], category))

print(catalog)
```

```python
category_to_product = {}
```

```python
for product, price, category in catalog:
    category_to_product.setdefault(category, []).append(product)

print(category_to_product)
```

```python
max_category = max(category_to_product, key=lambda x: len(category_to_product[x]))
print(max_category)
category_to_product[max_category]
```
