```python
# Task - 1 Basic Function: Price After Discount
```

```python
def calculate_discount(price, discount_percent):
    return price - (price * discount_percent / 100)


def apply_discount(price, discount_percent=5):
    
    if discount_percent > 60:
        discount_percent = 60
    
    return calculate_discount(price, discount_percent)

print("Price after discount:", apply_discount(1000, 10))
print("Price after discount (default):", apply_discount(500))

prices = [1000, 500, 200, 50]

def price_filter(p):
    return p > 100

def apply_default_discount(p):
    return apply_discount(p)

filtered_prices = list(filter(price_filter, prices))
discounted_prices = list(map(apply_default_discount, filtered_prices))

print("Filtered Prices:", filtered_prices)
print("Discounted Prices:", discounted_prices)
```

```python
# Task - 2 Recursive Function
```

```python
def factorial(n):
    if(n < 0):
        print("Enter a number greater than 0")
        return None

    if(n == 0 or n == 1):
        return 1

    else:
        return n * factorial(n-1)

print("Factorial of 5", factorial(5))
print("Factorial of 0", factorial(0))
print("Factorial of -3", factorial(-3))
```

```python
# Task - 3 Lambda Function: GST Calculator
```

```python
gst = lambda price: price + (price * 0.18)

print("GST: ", gst(100))

final_price = lambda price: (price + price * 0.18) - (price * 0.10)

print("Final price after applying discount", final_price(100))
```

```python
# Task - 6 Combined Utility Functions
```

```python
def process_prices(prices):
    discounted_prices = list(map(lambda p: p - (0.10 * p), prices))
    filtered_prices = list(filter(lambda p: p > 300, discounted_prices))
    
    return discounted_prices, filtered_prices

prices = [100, 500, 900, 50, 750]

discounted, filtered = process_prices(prices)

print("Discounted Prices:", discounted)
print("Filtered Prices (>300):", filtered)
```

```python
# Task - 7 Mini Problem: Menu using functions
```

```python
def add_price(prices_list, price):
    prices_list.append(price)


def get_average_price(prices_list):
    if len(prices_list) == 0:
        return 0
    return sum(prices_list) / len(prices_list)


def get_max_price(prices_list):
    if len(prices_list) == 0:
        return None
    return max(prices_list)


prices_list = []

while True:
    print("\nMenu")
    print("1 -> Add price")
    print("2 -> Show average price")
    print("3 -> Show highest price")
    print("q -> Quit")

    choice = input("Enter choice: ")

    if choice == "1":
        price = float(input("Enter price: "))
        add_price(prices_list, price)

    elif choice == "2":
        print("Average Price:", get_average_price(prices_list))

    elif choice == "3":
        print("Highest Price:", get_max_price(prices_list))

    elif choice == "q":
        print("Exiting program...")
        break

    else:
        print("Invalid choice")
```
