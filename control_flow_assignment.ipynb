```python
# Task - 1 Discount Rules
```

```python
order_amount = int(input("Enter amount: "))

if order_amount >= 2000:
    discount = 15
elif order_amount >= 1500 and order_amount < 2000:
    discount = 10
elif order_amount >= 1000 and order_amount < 1500:
    discount = 7
else:
    discount = 0

discount_amount = order_amount * discount / 100
sub_total = order_amount - discount_amount

tax = sub_total * 0.05
final_total = sub_total + tax

print("Orignal amount: ", order_amount)
print("Discount amount: ", discount_amount)
print("Final amount after applying discount: ", sub_total)
print("After tax amount: ", final_total)
```

```python
# Task - 2 Process Muliple orders
```

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_order_cnt = 0

print("Summary Table")

for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500 and order < 2000:
        discount = 10
    elif order >= 1000 and order < 1500:
        discount = 7
    else:
        discount = 0

    discount_amount = order * discount / 100
    final_total = order - discount_amount

    total_revenue += final_total

    if discount > 0:
        discounted_order_cnt += 1

    print(order, discount, final_total)
print("Total revenue", total_revenue)
print("number of order with discount", discounted_order_cnt)
```

```python
# Task - 3 User Menu
```

```python
orders = []

while True:
    print("Menu options: ")
    print("1 - Add order to a running list")
    print("2 - Show all orders and totals after applying discounts")
    print("q - Quit")

    choice = input("Enter your choice")

    if choice == "1":
        amount = int(input("Enter order amount: "))
        orders.append(amount)
        print("order added")

    elif choice == "2":
        if len(orders) == 0:
            print("No orders available")
            continue

        total_revenue = 0
        discounted_order_cnt = 0

        print("Summary Table")

        for order in orders:
            if order >= 2000:
                discount = 15
            elif order >= 1500 and order < 2000:
                discount = 10
            elif order >= 1000 and order < 1500:
                discount = 7
            else:
                discount = 0

            discount_amount = order * discount / 100
            total_amount = order - discount_amount
            

            total_revenue += total_amount

            if discount > 0:
                discounted_order_cnt += 1

            print(order, " ", discount, " ", total_amount)

            print("Total revenue: ", total_revenue)
            print("Number of orders with discount: ", discounted_order_cnt)

    elif choice == "q":
        print("Exiting program")
        break

    else:
        print("Invalid choice. Try again")
        continue
```

```python
# Task - 4 Loop control with conditions
```

```python
daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for day in daily:
    if day == -1:
        print("Corrupted data")
        break

    elif day == 0:
        print("Day with no sales")
        continue

    else:
        total_sales += day
        print("Running Total: ", total_sales)

    print("Final Total: ", total_sales)
```
